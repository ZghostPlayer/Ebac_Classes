# Probabilidade - Tarefa

```python
import numpy as np
from scipy import stats
from scipy.stats import norm, t
import matplotlib.pyplot as plt
```

1. Uma máquina embala ração para cachorro em pacotes de 15 kg, mas cada pacote produzido acaba tendo um peso aleatório, com distribuição Normal, com média 15kg, mas com desvio padrão 0,1 kg. Obtenha a probabilidade de que um pacote selecionado aleatoriamente dessa produção tenha mais que 15,2 kg.

```python
# Definição dos parâmetros da distribuição normal
media = 15  # Média do peso dos pacotes
desvio_padrao = 0.1  # Desvio padrão do peso dos pacotes
limite_superior = 15.2  # Valor para o qual queremos calcular a probabilidade

# Calculando a probabilidade acumulada até 15.2kg
probabilidade_acumulada = norm.cdf(limite_superior, loc=media, scale=desvio_padrao)

# Como queremos P(X > 15.2), fazemos 1 - P(X <= 15.2)
probabilidade_maior_que_15_2 = 1 - probabilidade_acumulada

# Exibir o resultado
print(f"A probabilidade de um pacote ter mais de 15,2 kg é de {probabilidade_maior_que_15_2:.4%}")
```

2. Da máquina descrita acima, obtenha o percentil de ordem 95% para o peso dos pacotes. Ou seja, o valor tal que a probabilidade de um pacote ter peso menor que este seja de 95%.

```python
percentil = 0.95  # Percentil desejado (95%)

# Calculando o percentil de ordem 95%
valor_percentil_95 = norm.ppf(percentil, loc=media, scale=desvio_padrao)

# Exibir o resultado
print(f"O percentil de 95% para o peso dos pacotes é {valor_percentil_95:.2f} kg")
```

3. Suponha duas variáveis aleatórias $Y_1$ e $Y_2$, a primeira com distribuição Normal Padrão (média=0, variância=1), a segunda com distribuição *t-studdent* com 5 graus de liberdade (também padrão). Calcule, para cada uma, a probabilidade de ser menor que cada um dos valores contidos no objeto $Y$ na célula abaixo:

```python
Y=np.linspace(-6, 6, 200)
# Seu código começa aqui

# Calculando a probabilidade acumulada para cada valor de Y na distribuição Normal Padrão
prob_Y1 = norm.cdf(Y, loc=0, scale=1)

# Calculando a probabilidade acumulada para cada valor de Y na distribuição t-Student com 5 graus de liberdade
prob_Y2 = t.cdf(Y, df=5)

# Exibir os primeiros valores como exemplo
print("Primeiros valores de Y:", Y[:5])
print("Probabilidade acumulada para Normal Padrão:", prob_Y1[:5])
print("Probabilidade acumulada para t-Student (5 GL):", prob_Y2[:5])
```

4. Faça uma visualização gráfica dos dados do item anterior.

```python
# Criando o gráfico
plt.figure(figsize=(8, 5))
plt.plot(Y, prob_Y1, label="Normal Padrão (μ=0, σ²=1)", linestyle="-", linewidth=2)
plt.plot(Y, prob_Y2, label="t-Student (GL=5)", linestyle="--", linewidth=2)

# Adicionando título e rótulos
plt.title("Distribuições Acumuladas: Normal Padrão vs. t-Student (5 GL)")
plt.xlabel("Valor de Y")
plt.ylabel("Probabilidade Acumulada")
plt.legend()
plt.grid(True)

# Exibir o gráfico
plt.show()
```

5. 3. Suponha duas variáveis aleatórias $Y_1$ e $Y_2$, a primeira com distribuição Normal Padrão (média=0, variância=1), a segunda com distribuição *t-studdent* com 5 graus de liberdade (também padrão). Calcule, para cada uma, o percentil (a inversa da *CDF*) para cada um dos valores do objeto *P* abaixo:

```python
P=np.linspace(0, 1, 20)
# Seu código começa aqui
# Calculando o percentil (inversa da CDF) para cada valor de P na Normal Padrão
percentil_Y1 = norm.ppf(P, loc=0, scale=1)

# Calculando o percentil (inversa da CDF) para cada valor de P na t-Student com 5 graus de liberdade
percentil_Y2 = t.ppf(P, df=5)

# Exibir os primeiros valores como exemplo
print("Valores de P:", P)
print("Percentis na Normal Padrão:", percentil_Y1)
print("Percentis na t-Student (5 GL):", percentil_Y2)
```

6. Faça um gráfico com os dados do item anterior.

```python
# Criando o gráfico
plt.figure(figsize=(8, 5))
plt.plot(P, percentil_Y1, marker='o', linestyle="-", linewidth=2, label="Normal Padrão (μ=0, σ²=1)")
plt.plot(P, percentil_Y2, marker='s', linestyle="--", linewidth=2, label="t-Student (GL=5)")

# Adicionando título e rótulos
plt.title("Percentis da Normal Padrão vs. t-Student (5 GL)")
plt.xlabel("Probabilidade Acumulada (P)")
plt.ylabel("Valor do Percentil")
plt.legend()
plt.grid(True)

# Exibir o gráfico
plt.show()
```
