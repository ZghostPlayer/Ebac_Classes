# Cálculo Numérico da Integral da Função $x^3$ entre 1 e 2

A integral definida da função $x^3$ no intervalo $[1, 2]$ pode ser aproximada utilizando o **método dos retângulos**. Nesse método, o intervalo é dividido em $n$ subintervalos iguais, e a área sob a curva é aproximada pela soma das áreas dos retângulos.

---

### Fórmula Geral

A área total é dada por:

$$
\text{Área Total} = \sum_{i=1}^n f(x_i) \cdot \Delta x
$$

Onde:  
- $\Delta x = \frac{b - a}{n}$ é a largura de cada retângulo.  
- $x_i$ é o ponto onde a função será avaliada (neste caso, o ponto médio de cada subintervalo).  

---

### Parâmetros Utilizados

- Função: $f(x) = x^3$  
- Limites de integração: $a = 1$, $b = 2$  
- Número de retângulos: $n = 1000$  

---

### Passo a Passo do Cálculo

1. **Divisão do Intervalo**:  

Dividimos o intervalo $[1, 2]$ em $n = 1000$ partes iguais:

$$
\Delta x = \frac{b - a}{n} = \frac{2 - 1}{1000} = 0.001
$$

2. **Cálculo dos Pontos Médios**:  

Os pontos médios dos retângulos são dados por:

$$
x_i = a + \left(i - \frac{1}{2}\right) \cdot \Delta x, \quad \text{para } i = 1, 2, \dots, n
$$

3. **Soma das Áreas**:  

Calculamos as áreas dos retângulos e somamos:

$$
\text{Área Total} \approx \sum_{i=1}^n f(x_i) \cdot \Delta x
$$

Como $f(x) = x^3$, temos:

$$
\text{Área Total} \approx \sum_{i=1}^n (x_i)^3 \cdot \Delta x
$$

---

### Resultado

Após realizar o cálculo com $n = 1000$, encontramos:

$$
\int_1^2 x^3 \, dx \approx 3.75
$$

O valor foi obtido com **1000 retângulos**, garantindo uma boa aproximação.

# Cálculo Analítico da Integral da Função $x^3$ entre 1 e 2

Queremos calcular a integral definida da função $x^3$ no intervalo $[1, 2]$:

$$
\int_1^2 x^3 \, dx
$$

---

### Passo 1: Calculando a Integral Indefinida

A integral de $x^3$ é calculada usando a regra da potência:

$$
\int x^n \, dx = \frac{x^{n+1}}{n+1} + C
$$

Portanto:

$$
\int x^3 \, dx = \frac{x^4}{4} + C
$$

---

### Passo 2: Avaliando a Integral Definida

Agora aplicamos os limites de integração $1$ e $2$:

$$
\int_1^2 x^3 \, dx = \left[\frac{x^4}{4}\right]_1^2
$$

Isso significa:

$$
\int_1^2 x^3 \, dx = \frac{2^4}{4} - \frac{1^4}{4}
$$

---

### Passo 3: Realizando os Cálculos

Primeiro, calculamos cada termo:

$$
\frac{2^4}{4} = \frac{16}{4} = 4
$$

$$
\frac{1^4}{4} = \frac{1}{4}
$$

Subtraindo:

$$
\int_1^2 x^3 \, dx = 4 - \frac{1}{4} = \frac{16}{4} - \frac{1}{4} = \frac{15}{4}
$$

---

### Resultado Final

A área calculada de forma analítica é:

$$
\int_1^2 x^3 \, dx = \frac{15}{4} = 3.75
$$
